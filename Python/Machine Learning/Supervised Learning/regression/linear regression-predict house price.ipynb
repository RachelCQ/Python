```python
# import liberaies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# import dataset
salary = pd.read_csv('C:\\Users\\Qing Chen\\Desktop\\dataset\\salary.csv')
X = salary.iloc[:,0].values.reshape(-1,1)
y = salary.iloc[:,1].values.reshape(-1,1)
X
```

```python
#see how many kind of years
sns.distplot(salary['YearsExperience'],kde=False,bins=10)
```

```python
sns.countplot(y='YearsExperience',data=salary)
```

```python
# visualize the data
sns.barplot(x = 'YearsExperience', y='Salary',data=salary)
```

```python
sns.heatmap(salary.corr())
```

```python
# split the data into training and testing set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 1/3, random_state = 0)
```

```python
# fit simple linear regression to the training dataset
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
# predct the test set results
y_pred = lr.predict(X_test)
y_pred
```

```python
# visulize the test set results:
plt.scatter(X_test,y_test,color='blue')
# plt.scatter(X_test,lr.predict(X_test),color='blue')
plt.plot(X_test,lr.predict(X_test),color='red')
plt.title('salary vs experience(test,set)')
plt.xlabel('years of experience')
plt.ylabel('salary')
plt.show()
```

```python
# caculate the residuals
from sklearn import metrics
print('MAE:',metrics.mean_absolute_error(y_test,y_pred))
print('MSE:', metrics.mean_squared_error(y_test,y_pred))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

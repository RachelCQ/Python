```python
import pandas as pd
```

## Open the file "fashion.csv" using pandas.

```python
# read fashion data
df = pd.read_csv('F:\\2019Spring\\Text Analytics\\homework\\ASN3\\fashion.csv')
df.head()
```

```python
# get review column
review_df = df['review text']
review_df.head()
```

## Generate polarity scores - positive, negative, neutral - for each review using NLTK*.

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer # Vader

res = []
sid = SentimentIntensityAnalyzer()
for sentence in review_df:
    ss = sid.polarity_scores(sentence)
    row = []
    for k in ss:
        row.append(ss[k])
    res.append(row)
res
```

```python
# trun res(list) into matrix
data = pd.DataFrame(res,columns= ['Negative','Neutral','Positive','comp'])
data.head()
```

## build columns: 1. ID, 2. Positive, 3. Negative, and 4. Neutral scores

```python
# delet 'comp' column
data1 = data.drop('comp',axis=1)
```

```python
# change the order of columns
data2 = data1.loc[:,['Positive','Negative','Neutral']]
data2.head()
```

```python
# add an id column on the left
import numpy as np
id_column = pd.DataFrame(np.arange(1,len(data2)+1), columns=['ID'])
id_column.head()
```

```python
# combine ID and data
new_data = pd.concat([id_column,data2],axis=1)
new_data.head()
```

## Save the output as "sentiment_output.csv".

```python
new_data.to_csv('F:\\2019Spring\\Text Analytics\\homework\\ASN3\\sentiment_output.csv', index = None)
```

## Merge the output file with the original input file.

```python
merge_orginal = pd.concat([df,new_data],axis = 1)
merge_orginal.head()
```

## Summarize the sentiment scores for each of the four cities – New York, Paris, Milan, and London.

```python
city_scoredf = pd.DataFrame(merge_orginal[['location','Positive','Negative','Neutral']])
city_scoredf = pd.concat([city_scoredf,data['comp']],axis = 1)
# for city in city_scoredf['location']:
#     if city == 'NEW YORK':
city_scoredf.head()
        
```

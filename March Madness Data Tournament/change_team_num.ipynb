# Exchange team1 and team2 information to fit for machine learning

### Problem Statement: team1 always has higher score than team2 in original data, which results in the outcome column is all "Win" based on "team1_Score >team2_score"

### Solution: Exchange half of team1 and team2 information

```python
import pandas as pd
import numpy as np
from copy import deepcopy

df = pd.read_csv('F:\\2019Spring\\march madness\\New folder\\NCAA_Tourney_2002_2018.csv')
df.head()
```

```python
alist = list(df.columns)
newlist = deepcopy(alist)
alist
```

```python
for i in range(len(newlist)):
    if newlist[i].startswith('team1'):
        newlist[i] = newlist[i].replace('team1','team2')
    elif newlist[i].startswith('team2'):
        newlist[i] = newlist[i].replace('team2','team1')
    else:
        print(newlist[i])
newlist
```

```python
ncdf = pd.read_csv('F:\\2019Spring\\march madness\\New folder\\NCAA_Tourney_2002_2018.csv',names=newlist)
newdf= ncdf.drop(index = 0)
newdf = newdf.reset_index(drop=True)
newdf.head()
```

```python
order = alist
betterdf = newdf[alist]
```

```python
gameid = betterdf['game_id']
newid = []
for i in gameid:
    idlist = i.split('-')
    t1 = idlist[1]
    t2 = idlist[2]
    idlist[1] = t2
    idlist[2] = t1
    newid.append('-'.join(idlist))

print(len(newid))
```

```python
id = pd.DataFrame(newid,columns=['game_id'])
id
```

```python
be2df = betterdf.drop(['game_id'],axis=1)
be2df.head()
len(be2df)
```

```python
best =pd.concat([be2df,id],axis=1)
best
```

```python
wholedf =pd.concat([df,best],axis=0)
wholedf = wholedf.reset_index(drop=True)
wholedf
```

```python
for i in range(555,1666):
    wholedf= wholedf.drop(index = i)

wholedf
```
